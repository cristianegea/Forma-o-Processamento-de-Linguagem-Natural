```python
# Importação das bibliotecas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
# Importação do arquivo
from google.colab import files
uploaded = files.upload()
```

```python
# Criação um objeto dataframe
spam = pd.read_csv('spam.csv')
spam.head()
```

```python
# Exibição da dimensão do dataframe
spam.shape
```

```python
# Exibição do número de elementos de cada categoria
spam.Category.value_counts()
```

```python
# Separação das variáveis dependente e independentes
previ = spam.Message
classe = spam.Category
```

```python
# Vetorização (modelo ML não consegue processar textos)
vetorizador = TfidfVectorizer()
previsores = vetorizador.fit_transform(previ)
previsores.shape
```

```python
# Exibição de parte da matriz criada por meio da vetorização
vetorizador.get_feature_names_out()[10:100]
```

```python
# Separação dos dados em treinbo e teste
X_treinamento, X_teste, y_treinamento, y_teste = train_test_split(previsores, classe, test_size = 0.3)
```

```python
# Criação do modelo de classificação
floresta = RandomForestClassifier(n_estimators=500)
floresta.fit(X_treinamento, y_treinamento)
```

```python
# Previsão das classes
previsoes = floresta.predict(X_teste)
```

```python
# Geração da matriz de confusão
confusion_matrix(y_teste, previsoes)
```

```python
# Acurácia do modelo
accuracy_score(y_teste, previsoes)
```

```python
# Relatório com resumo das métricas de avaliação do desempenho do modelo
print(metrics.classification_report(y_teste, previsoes))
```
